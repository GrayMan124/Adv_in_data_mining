We start with the probality in the case of coins in the buckets. 
Let:
- N number of buckets

For a specific bucket (let's say number 7) the probability of the first coin going into that bucket is 1/N, so the probabilty of the bucket 7 being empty after the first throw is 1-1/N. 

For the second throw it will be (1-1/N)^2 because, now we need to miss two times 

for the k-th throw the probability of it being empty is (1-1/N)^k, again we would have to miss the bucket 7 all k times.

Now that we have the probability of a specific bucket being empty after k throws, we can calculate to probabilty of it not being empty after k throws which will be 1-(1-1/N)^k

If we want to get the expected number of buckets not being empty, then we just have to sum the probabilties of every bucket: 

$\sum_{i=1}^{N} 1-(1-1/N)^k$

Here is the code that computes the value of the sum written above for 6000 buckets, and 7000 coins (or, 7000 integers that are lower than 6000)

```python
N = 6000
k = 7000
```

```python
expected = np.sum(np.ones([6000])*(1-np.power(1-1/6000,7000)))
expected
```

To verify the forumla, we have tried different combinations of the numbers of buckets, and throws (in other words number the size of the array of integers, and the range of the integers) 

The code here presents the results

```python
def expected(N,k):
    return np.sum(np.ones([N])*(1-np.power(1-1/N,k)))
```

```python
N = 10
k = 10

tmp = []

for i in range(100):
    random_array = np.random.randint(N,size = k )
    tmp.append(np.unique(random_array).shape[0] )
# print(tmp)
tmp = np.array(tmp)
print("unique values in the array: "f'{np.mean(tmp)}')
print("Expected unique values from the formula: "f'{expected(N,k)}')
# expected(N,k)
```
